```python
import requests
import re
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
def scrap(id_start, id_end):
    patients = pd.DataFrame(columns=['id', 'Name', 'url'])
    patients['id'] = patients['id'].astype(int)
    for id in range(id_start, id_end):
        url = f'https://{id}.portal.athenahealth.com/'
        response = requests.get(url)
        html = response.text
        soup = bs(html, 'html.parser')
        try:
            title = soup.find_all('h1')[1].text
            reg = r'Welcome to the (.+?) Patient Portal'
            resultado = re.search(reg, title)
            if resultado:
                patients = pd.concat([patients, pd.DataFrame({'id': [id], 'Name': [resultado.group(1)], 'url': [url]})], ignore_index=True)
        except Exception:
            continue
    patients.to_excel('patients_scraping.xlsx', index=False)

    return patients
```

```python
prueba = scrap(12600, 12650)
prueba
```
